```python
 import matplotlib.pyplot as plt

# Data for IVA_G
k_g = [12, 13, 14, 15, 16, 17, 18, 19]
f1_score_g = [0.736752, 0.674419, 0.746183, 0.747982,0.681779, 0.736508, 0.753292, 0.753292]

# Data for IVA_SPICE
k_spice = [12, 13, 14, 15, 16, 17, 18, 19]
f1_score_spice = [0.618434, 0.706282, 0.746183, 0.738812,0.638596, 0.736508, 0.627016, 0.732738]

# Plotting
plt.figure(figsize=(8, 5))

plt.plot(k_g, f1_score_g, marker='o', linestyle='-', color='b', label='IVA_G')
plt.plot(k_spice, f1_score_spice, marker='s', linestyle='--', color='r', label='IVA_SPICE')

plt.title('F1 Scores for IVA_G and IVA_SPICE')
plt.xlabel('k')
plt.ylabel('F1 Score')
plt.xticks(k_g)  # Set x-axis ticks to be consistent with both datasets
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
